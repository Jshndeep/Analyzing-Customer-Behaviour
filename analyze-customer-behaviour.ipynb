# **Customer Segmentation**

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
from sklearn.preprocessing import StandardScaler

segmentation_data = df[['QTY', 'BILL_AMT']]
scaler = StandardScaler()
segmentation_data_scaled = scaler.fit_transform(segmentation_data)

# Determine the optimal number of clusters using the Elbow Method
wcss = []  # Within-Cluster-Sum-of-Squares
for i in range(1, 11):
    kmeans = KMeans(n_clusters=i, init='k-means++', max_iter=300, n_init=10, random_state=0)
    kmeans.fit(segmentation_data_scaled)
    wcss.append(kmeans.inertia_)

# Plot the Elbow Method graph
plt.plot(range(1, 11), wcss)
plt.title('Elbow Method for Optimal k')
plt.xlabel('Number of Clusters (k)')
plt.ylabel('WCSS')  # Within-Cluster-Sum-of-Squares
plt.show()
```

```python
# Choose the optimal k and fit the KMeans model
optimal_k = 3  # Determine from the Elbow Method
kmeans = KMeans(n_clusters=optimal_k, init='k-means++', max_iter=300, n_init=10, random_state=0)
df['Cluster'] = kmeans.fit_predict(segmentation_data_scaled)

# Visualize the clusters
plt.scatter(df['QTY'], df['BILL_AMT'], c=df['Cluster'], cmap='viridis')
plt.title('Customer Segmentation')
plt.xlabel('Sales Unit')
plt.ylabel('Bill Amount')
plt.show()
```
